# Un tutorial de matplotlib
Matplotlib es una paquetería de Python que permite la visualización de datos.

Por lo general se utiliza extensamente en conjunto con el módulo NumPy.


# Importando los módulos
Por el momento usaremos el submódulo `pyplot` para la visualización de datos

```python
import numpy as np
import matplotlib.pyplot as plt
```

Las declaraciones anteriores nos permiten acceder a NumPy a Matplotlib mediante los alias `np` y `plt`, respectivamente.

# Diferentes tipos de gráficos
## Gráfica de una línea recta

```python
x = np.arange(10)
y_1 = 3 * x - 1
y_2 = -2 * x + 5

plt.plot(x, y_1, c='red', label='y = 3x - 1')
plt.plot(x, y_2, c='blue', label='y = -2x + 5')
plt.legend()

plt.xlabel('x')
plt.ylabel('f(x)')

plt.savefig('matplotlib_lines.png', bbox_inches='tight', pad_inches=0.03,  transparent=False, dpi=300)
plt.show()
```

## Gráfica de funciones matemáticas más complejas

```python
x = np.linspace(0, 4*np.pi, 100)
plt.plot(x, np.sin(x))
plt.show()
```

## Gráfica de histograma

```python
#- Distribución normal
x = np.random.normal(100, 20, 100)

#- Dibujar el histograma
plt.hist(x)

#- Añadir etiquetas a los ejes
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

#- Muestra la figura
plt.show()
```

## Gráfica tipo "scatter"

```python
x_data = np.random.random(100)
y_data = np.random.random(100)

plt.scatter(x_data, y_data)
plt.show()
```

## Gráfica con barras de error

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

y_error = 0.5*np.random.random(100)

plt.errorbar(x, y, yerr=y_error)
plt.show()
```

## Varios páneles de gráficas

```python
fig = plt.figure(figsize=(10, 5))

#- Subplot: 1 fila, 2 columnas, posición 1
ax1 = fig.add_subplot(121)
ax1.plot(x, y)

#- Subplot: 1 fila, 2 columnas, posición 2
ax2 = fig.add_subplot(122)
ax2.errorbar(x, y, yerr=y_error)

plt.show()
```

## Gráfica de arreglos de dos dimensiones

```python
matrix = np.random.random((100, 100))
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

```python
x = np.linspace(1, 20, 100)
y = np.linspace(1, 25, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
plt.contour(X, Y, Z)
plt.show()
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)  
```

```python
#- Valores de x e y
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

#- Malla de x e y
X, Y = np.meshgrid(x, y)

#- Malla de valores de z
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
plt.show()
```

```python
#- Definir la cuadrícula de puntos
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
```

```python
#- Media
mu_x, mu_y = 0, 0  

#- Desviación estándar
sigma_x, sigma_y = 1, 1  
```

```python
Z = (1 / (2 * np.pi * sigma_x * sigma_y)) * np.exp(-((X - mu_x)**2 / (2 * sigma_x**2) + (Y - mu_y)**2 / (2 * sigma_y**2)))
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(label='Densidad de probabilidad')
plt.title('Distribución Gaussiana en 2D')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
